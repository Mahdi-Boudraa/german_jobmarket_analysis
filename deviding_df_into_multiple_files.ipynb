```python
import re

def clean_german_location(location):
    # Convert to string if not already
    location = str(location).strip()
    
    # Common normalizations (keep German names but standardize variations)
    replacements = {
        'muenchen': 'MÃ¼nchen',
        'munich': 'MÃ¼nchen',
        'koeln': 'KÃ¶ln',
        'cologne': 'KÃ¶ln',
        'nuernberg': 'NÃ¼rnberg',
        'nuremberg': 'NÃ¼rnberg',
        'frankfurt am main': 'Frankfurt am Main',
        'frankfurt/main': 'Frankfurt am Main',
        'frankfurt a. m.': 'Frankfurt am Main',
        'frankfurt a.m.': 'Frankfurt am Main',
        'hannover': 'Hannover',  # Keep German spelling
        'duesseldorf': 'DÃ¼sseldorf'
    }
    
    # Remove anything in parentheses, after commas, or special characters
    loc = re.split(r'[,(;]', location)[0].strip()
    
    # Remove postal codes and numbers
    loc = re.sub(r'\d{4,5}\s*\w*', '', loc).strip()
    
    # Remove "DE", "Deutschland", "Germany" etc.
    loc = re.sub(r'(DE,?|deutschland|germany|,?\s*Germany|,?\s*Deutschland)', '', loc, flags=re.IGNORECASE).strip()
    
    # Apply replacements while preserving German names
    loc_lower = loc.lower()
    for old, new in replacements.items():
        if old in loc_lower:
            loc = new
            break
    
    # Special handling for major German cities
    major_cities = ['Berlin', 'MÃ¼nchen', 'Hamburg', 'Frankfurt am Main', 'KÃ¶ln', 
                   'DÃ¼sseldorf', 'Stuttgart', 'NÃ¼rnberg', 'Hannover', 'Leipzig']
    
    for city in major_cities:
        if city.lower() in loc.lower():
            return city
    
    # Capitalize properly (respect German umlauts and ÃŸ)
    if len(loc) > 0:
        loc = loc[0].upper() + loc[1:]
    
    return loc
```

```python
df_general["location"] = df_general.location.apply(clean_german_location)
```

```python
df_general.to_csv("general.csv")
```
